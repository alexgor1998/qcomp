```python
import numpy as np
from qiskit import *
from math import sqrt, pi
with open('Qconfig_IBMQ_experience.py', 'r') as token_file:
    token = token_file.read()
IBMQ.save_account(token, overwrite = True)
IBMQ.load_account()
my_provider = IBMQ.get_provider()
```

```python
qreg_q = QuantumRegister(2, 'q')
creg_c = ClassicalRegister(2, 'c')
circuit = QuantumCircuit(qreg_q, creg_c)
# also see note on Google drive
circuit.reset(qreg_q[0])
circuit.reset(qreg_q[1])
#preparing H^2|00> state 
circuit.h(qreg_q[0])
circuit.h(qreg_q[1])

circuit.barrier(qreg_q[0], qreg_q[1])

# an oracle for a = |01>
#circuit.x(qreg_q[1])
#circuit.cz(qreg_q[0], qreg_q[1])
#circuit.x(qreg_q[1])

# another oracle example — for |00>
circuit.x(qreg_q[0])
circuit.x(qreg_q[1])
circuit.cz(qreg_q[0], qreg_q[1])
circuit.x(qreg_q[0])
circuit.x(qreg_q[1])

circuit.barrier(qreg_q[0], qreg_q[1])
# applying -W
circuit.h(qreg_q[0])
circuit.h(qreg_q[1])

circuit.x(qreg_q[0])
circuit.x(qreg_q[1])
circuit.cz(qreg_q[0], qreg_q[1])
circuit.x(qreg_q[0])
circuit.x(qreg_q[1])

circuit.h(qreg_q[0])
circuit.h(qreg_q[1])

circuit.barrier(qreg_q[0], qreg_q[1])

circuit.measure(qreg_q[0], creg_c[0])
circuit.measure(qreg_q[1], creg_c[1])


backend=Aer.get_backend('qasm_simulator')
jobIBM=execute(circuit,backend=backend,shots=1024)
resultSim=jobIBM.result()

backend  = my_provider.get_backend('ibmq_ourense')
jobIBM=execute(circuit,backend=backend,shots=1024)
resultExp=jobIBM.result()

from qiskit.tools.visualization import plot_histogram
#plot_histogram([resultSim.get_counts(circuit)])
plot_histogram([resultExp.get_counts(circuit),resultSim.get_counts(circuit)])
```

```python
qreg_q = QuantumRegister(2, 'q')
creg_c = ClassicalRegister(2, 'c')
circuit = QuantumCircuit(qreg_q, creg_c)

#let's try to apply more than 1 iteration
circuit.reset(qreg_q[0])
circuit.reset(qreg_q[1])
#preparing H^2|00> state 
circuit.h(qreg_q[0])
circuit.h(qreg_q[1])

ctr = 3
for i in range(ctr):
    # an oracle for a = |11>
    circuit.cz(qreg_q[0], qreg_q[1])

    # applying -W
    circuit.h(qreg_q[0])
    circuit.h(qreg_q[1])

    circuit.x(qreg_q[0])
    circuit.x(qreg_q[1])
    circuit.cz(qreg_q[0], qreg_q[1])
    circuit.x(qreg_q[0])
    circuit.x(qreg_q[1])

    circuit.h(qreg_q[0])
    circuit.h(qreg_q[1])

circuit.measure(qreg_q[0], creg_c[0])
circuit.measure(qreg_q[1], creg_c[1])


backend=Aer.get_backend('qasm_simulator')
jobIBM=execute(circuit,backend=backend,shots=1024)
resultSim=jobIBM.result()

backend  = my_provider.get_backend('ibmq_ourense')
jobIBM=execute(circuit,backend=backend,shots=1024)
resultExp=jobIBM.result()

from qiskit.tools.visualization import plot_histogram
#plot_histogram([resultSim.get_counts(circuit)])
plot_histogram([resultExp.get_counts(circuit),resultSim.get_counts(circuit)])

#we got even distribution on 2 and  3 repetitions
#this is due do |phi> being not much orthogonal to |a> and state being exaclty a after 1 iteration
#also see notes
```
